```python
# first want to load a single file
import os
import json

directory = '../data/raw/tests_male_json'
files = sorted([f for f in os.listdir(directory) if f.endswith('.json')])
last_file_path = os.path.join(directory, files[-1])


print(f"Loaded: {len(files)} Test Matches Available")
```

```python
# look at one file
last_file_path = os.path.join(directory, files[-1])

with open(last_file_path, 'r') as f:
    data = json.load(f)

match_info = data['info']

# extract start date, team_type, match_type, gender, match_id, team1, team2 
start_date = match_info['dates'][0]
team_type = match_info['team_type']
match_type = match_info['match_type']
gender = match_info['gender']
match_id = last_file_path.split('/')[-1].split('.')[0]
team1 = match_info['teams'][0]
team2 = match_info['teams'][1]

print(start_date, team_type, match_type, gender, match_id, team1, team2)

```
